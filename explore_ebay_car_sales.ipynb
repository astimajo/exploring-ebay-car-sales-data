# Data Cleaning

# Cleaning Column Names
Below we are going to convert the dataframe column names from camel case to snake case. We will also reword some column names to be more descriptive.

```python
# print array of columns for overview
autos.columns
```

```python
# Cleaning column names to remove snake case and etc.
# Columns had to be manually cleaned due to camel casing 
# some colums also requires rephrasing.
autos_columns = autos.columns
autos_columns = autos_columns.str.replace('yearOfRegistration','registration_year')
autos_columns = autos_columns.str.replace('monthOfRegistration','registration_month')
autos_columns = autos_columns.str.replace('notRepairedDamage','unrepaired_damage')
autos_columns = autos_columns.str.replace('dateCreated','ad_created')
autos_columns = autos_columns.str.replace('dateCrawled','date_crawled')
autos_columns = autos_columns.str.replace('offerType','offer_type')
autos_columns = autos_columns.str.replace('vehicleType','vehicle_type')
autos_columns = autos_columns.str.replace('powerPS','power_ps')
autos_columns = autos_columns.str.replace('fuelType','fuel_type')
autos_columns = autos_columns.str.replace('nrOfPictures','nr_of_pictures')
autos_columns = autos_columns.str.replace('postalCode','postal_code')
autos_columns = autos_columns.str.replace('lastSeen','last_seen')

# reassign back to df.columns
autos.columns = autos_columns #cleaned version

# Print five samples to check current state of autos.
autos.head()
```

The changes here are mainly focused on converting the previous camel cased column names to snake case which is the recommended case-style for python programs.

Other changes include rephrasing the column names to enhance the understandability of the Dataset. As we might observe, the changes were done in a manually keyed manner. This is mainly because the dataset has no uniform changes that can be applied and each name-change is unique.

snake-case conversion might have been done differently but we must first create a special function to achieve that as there are no standard modules that can achieve such. But that can be improved some other time since we have already achieved the column cleaning stage.

# Dealing with incorrect Registration Year Data

We shall start fixing this column below

```python
# We will have to determine what are the acceptable highest
# and lowest year values before we can proceed.
# We will detect upper and lower bounds first.

registration_year = autos['registration_year']
registration_year.value_counts().sort_index()
```

We can decide now that the most reasonable upper bound would be `2016`, anything beyond 2016 would be inaccurate data.

since cars were invented in 1886 it is impossible to see cars registered in 1800, we can then conclude the earliest possible date would be `1910`. This I believe is reasonable since antique ford vehicles do exist.

```python
# remove the data points outside the upper and lower bounds.
x = 1910
y = 2016

autos = autos[(autos['registration_year'] >= x) & (autos['registration_year'] <= y)]

# check the registration_year column to see effects
autos['registration_year'].value_counts().sort_index()
```

Effects of the Boolean indexing is now effected and we have now removed irrelevant data from the registration year column.

# Exploring Price by Brand

In this section we will explore car brands in the dataset

```python
autos_brand = autos['brand'].unique()
autos['brand'].value_counts().head(6)
```

The selected brands for aggregation are the top 6 brands with the most presence in the dataset.

In this portion we will compute the average price per brand.

```python
selected_brands = autos['brand'].value_counts().head(6)
brand_ave_price = {}

for item in selected_brands.index:
    selected_brands = autos[autos['brand'] == item]
    ave_price = selected_brands['price'].mean().astype(int)
    brand_ave_price[item] = ave_price
```

```python
print(brand_ave_price)
```

It looks like the most expensive car brand is audi, This is followed by bmw then mercedes-benz.

The results may vary from analysis to analysis due to the decisions to exclude or include certain datapoints in the cleaning process.

Sonstige_autos may seem like a brand but this isn't, This means other brands but in German.

# Storing Aggregate Data in a Dataframe

```python
brand_ave_series = pd.Series(brand_ave_price)
brand_ave_df = pd.DataFrame(brand_ave_series, columns = ['mean_price'])
```

```python
brand_ave_df.sort_values(['mean_price'], ascending = False)
```

We will now try to determine if there is a correlation between mileage and price for our average prices.

We will aggregate two series into a dataframe to perform a sort of correlation check between the 2 series' datapoints. We will use mean mileage for the second series.

```python
selected_brands = autos['brand'].value_counts().head(6)
brand_ave_mileage = {}

for item in selected_brands.index:
    selected_brands = autos[autos['brand'] == item]
    ave_mileage = selected_brands['odometer_km'].mean().astype(int)
    brand_ave_mileage[item] = ave_mileage
```

```python
print(brand_ave_mileage)
```

```python
brand_mileage_series = pd.Series(brand_ave_mileage)
brand_ave_df['mean_mileage'] = brand_mileage_series
```

```python
brand_ave_df.sort_values(['mean_price'], ascending = False)
```

```python
brand_ave_df.sort_values(['mean_mileage'], ascending = False)
```
